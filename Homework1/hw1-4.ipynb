```python
from cvxpy import Variable, quad_form, Minimize, Problem
import numpy as np
```

```python
x = Variable(1, name='x')
```

```python
# constraints
# P1 = np.array(np.mat('1. 0.5; 0.5 1.'))
# f1 = quad_form(x, P1)
# f2 = 3.*x[0]+2.*x[1]
# constraints = [f1 <= 3., f2>= 3.]
constraints = [x[0]**2 - 6*x[0] + 8 <= 0]
```

It's convex, yey



```python
# objective
# P0 = np.array(np.mat('1. 0.; 0. 1.'))
# f0 = quad_form(x, P0)
# obj = Minimize(f0)
obj = Minimize(x[0]**2 + 1)
```

```python
# solve
prob = Problem(obj, constraints)
print('solve:', prob.solve())
print('status:', prob.status)
print('optimal value =', prob.value)
print('x* =', x.value)
print('dual values:', [c.dual_value for c in constraints])
```
